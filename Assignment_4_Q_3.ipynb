```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 10000
l = 1
d = 2

# Initialize counters
crosses = 0
pi_estimates = []

# Set seed for reproducibility
np.random.seed(42)

# Run the Monte Carlo simulation
for i in range(1, N + 1):
    x = np.random.uniform(0, d / 2)

    theta = np.random.uniform(0, np.pi / 2)

    if (l / 2) * np.sin(theta) >= x:
        crosses += 1

    if crosses > 0:
        pi_est = (2 * i * l) / (d * crosses)
        pi_estimates.append(pi_est)
    else:
        pi_estimates.append(None)

# Plot the convergence of π estimate
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, label='Estimated π')
plt.axhline(y=np.pi, color='red', linestyle='--', label='True π (3.1416)')
plt.xscale('log')
plt.xlabel("Number of needle drops (log scale)")
plt.ylabel("Estimated π")
plt.title("Buffon's Needle: Monte Carlo Estimation of π")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Final estimate and analysis
final_pi = pi_estimates[-1]
print("\n")
print(f"Estimated value of π after {N} drops: {final_pi:.5f}")
print("\n")
print("1. Random seed variation: Using a different seed gives slightly different results.")
print("2. Sample size: Larger N leads to better convergence .")
print("3. Angle resolution: sin(θ) is sensitive near 0 and π/2; poor sampling here increases error.")
print("4. Early trials may fluctuate heavily due to low crossing counts.")
```
