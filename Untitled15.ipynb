```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1)
def simulate_wiener(T=1, N=1000):
    dt = T / N  # Time step size
    times = np.linspace(0, T, N + 1)  # Time grid from 0 to T
    W = [0]  # Initial value W(0) = 0
    for _ in range(N):
        dW = np.random.normal(0, np.sqrt(dt))  # Normal increment
        W.append(W[-1] + dW)  # Update W(t) by adding increment
    return times, W

# Set total time and number of steps
T = 1
N = 1000
# Generate one Wiener process path
times, W = simulate_wiener(T, N)

# Plot the result
plt.plot(times, W)
plt.title("Q1: Wiener Process Path")
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.grid(True)
plt.show()
```

```python
# 2)
def simulate_gbm(S0=1, alpha=0.1, sigma=0.2, T=1, N=1000, paths=5):
    dt = T / N  # Time step
    t = np.linspace(0, T, N + 1)  # Time grid
    all_paths = []  # Store all simulated GBM paths

    for _ in range(paths):
        W = [0]  # Start Brownian motion at 0
        for _ in range(N):
            dW = np.random.normal(0, np.sqrt(dt))  # Generate Brownian increment
            W.append(W[-1] + dW)  # Accumulate increments
        # Apply the GBM formula
        S = S0 * np.exp(np.array(W) * sigma + (alpha - 0.5 * sigma**2) * t)
        all_paths.append(S)  # Add path to list

    return t, all_paths

# Simulate and plot 5 GBM paths
t, gbm_paths = simulate_gbm()
for path in gbm_paths:
    plt.plot(t, path)
plt.title("Q2: Geometric Brownian Motion - 5 Paths")
plt.xlabel("Time")
plt.ylabel("S(t)")
plt.grid(True)
plt.show()
```

```python
#3)
# Theoretical result for Brownian motion: E[W_s * W_t] = min(s, t)

s = 0.4
t = 0.7
samples = 10000  # Number of simulations

# Generate W_s and W_t using their distributions
Ws = np.random.normal(0, np.sqrt(s), samples)
Wt = np.random.normal(0, np.sqrt(t), samples)

# When s < t, the correlation comes from W_s alone
# So E[W_s * W_t]  E[W_s^2] = Var(W_s) = s
print("Q3: Empirical E[Ws * Wt] ", np.mean(Ws * Ws), "Theoretical:", min(s, t))
```

```python
# 4)

s = 0.5
t = 1.0
N = 10000

# Simulate W_s ~ N(0, s)
W_s = np.random.normal(0, np.sqrt(s), N)
# Generate W_t = W_s + independent increment from N(0, t-s)
W_t = W_s + np.random.normal(0, np.sqrt(t - s), N)

# Difference should follow N(0, t-s)
diff = W_t - W_s
print("Q4: Mean of Wt - Ws:", np.mean(diff))
print("Variance of Wt - Ws:", np.var(diff))
```

```python
#5)
# The expected future value given current info is the current value
# Since W_t = W_s + independent noise, the expectation of the increment is zero

W_s = np.random.normal(0, np.sqrt(s), N)
W_t = W_s + np.random.normal(0, np.sqrt(t - s), N)

# The conditional expectation of the increment should be zero
cond_exp = np.mean(W_t - W_s)
print("Q5: E[Wt - Ws | Fs] ", cond_exp, "Expected: 0  Brownian motion is a martingale")
```
